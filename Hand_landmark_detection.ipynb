```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image_path ='./asl_datbase/asl_alphabet_test/asl_alphabet_test/B_test.jpg'
```

```python
image_tf_size = [256, 256]
# Image preprocessing
# tf.enable_eager_execution()

cap= cv2.VideoCapture(0)
while True:
    _,frame=cap.read()
#     print(frame.shape)
#     hand_img = tf.io.read_file(image_path)
#     print(frame.shape)

#     hand_img = tf.image.decode_jpeg(hand_img, channels=3)
#     print(hand_img.shape)

    hand_img = tf.image.convert_image_dtype(frame, tf.float32)
    hand_img = tf.image.resize(hand_img, image_tf_size)
    hand_img = hand_img.numpy()
    input_img = tf.expand_dims(hand_img, 0)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    # Load TFLite model and allocate tensors.
    # interpreter = tf.lite.Interpreter(model_path='PATH_TO/mediapipe/mediapipe/models/hand_landmark.tflite')
    interpreter = tf.lite.Interpreter(model_path='./hand_tracking_using_mediapipe_models/models/hand_landmark.tflite')
    interpreter.allocate_tensors()

    # Get input and output tensors.
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    # Test model on random input data.
    print(output_details)
    interpreter.set_tensor(input_details[0]['index'], input_img)

    interpreter.invoke()

    # The function `get_tensor()` returns a copy of the tensor data.
    # Use `tensor()` in order to get a pointer to the tensor.
    output_data = interpreter.get_tensor(output_details[0]['index'])

    coordinates = np.split(output_data[0], 21)
    print(output_data)

    for c in coordinates:
        image = cv2.circle(hand_img, (c[0],c[1]), radius=0, color=(0, 0, 255), thickness=2)

        print(c)

    cv2.imshow("image",image)
    plt.show()
```

```python
cap.release()
cv2.destroyAllWindows()
```
